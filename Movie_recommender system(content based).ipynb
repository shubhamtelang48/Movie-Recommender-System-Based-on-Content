```python
# we want this string conver into list

import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert) 
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# we need only first 3 cast of movies for recommender

def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview'][0]
```

```python
# because overview column is in string format we have to convert it into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.sample(5)
```

we want ['Matt Damon']=['MattDamon']
      ['Sam Worthington]=['SamWorthington]

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
# add this 5 column into one cloumn

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new_df.head()
```

```python
# we used this function to convert list into string

new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):               # stem will convert similar words into root words 
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        return ''.join(y)

new_df['tags']=new_df['tags'].apply(stem)

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
 vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
list(enumerate(similarity[1]))  # enumeratate function used for to not change the index of movie
```

```python
# It will give index of given movie 

new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
# It will give the sorted list of movie which is similar to given movie

sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```

```python
recommend('Avatar')
```

```python
recommend('John Carter')
```
